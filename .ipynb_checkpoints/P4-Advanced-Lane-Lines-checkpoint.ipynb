```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
import pickle
from moviepy.editor import VideoFileClip
from IPython.display import HTML
from scipy import signal

%matplotlib inline

global left_fit_prev
global right_fit_prev
left_fit_prev = np.array([0,0,0])
right_fit_prev = np.array([0,0,0])

global color_left
global color_right

color_left = (255,255,255)
color_right = (255,255,255)

global mask_init
mask_init = 0

# enable diagnostics view of the pipeline
global enable_diag
enable_diag = True

# Define size for sliding windows
window_size = 60

# Define color thresholds to extract white and yellow masks
hls_white_low = np.array([0, 180, 0])
hls_white_high = np.array([255, 255, 50])

hsv_yellow_low = np.array([0, 100, 120])
hsv_yellow_high = np.array([120, 255, 255])
```

```python
def get_obj_img_points (cal_imgs, nx, ny):
    objp = np.zeros((ny*nx,3), np.float32)
    objp[:,:2] = np.mgrid[0:nx, 0:ny].T.reshape(-1,2)
    # Arrays to store object points and image points from all the images.
    objpoints = []
    imgpoints = []
    for img in cal_imgs:
        # Convert to grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Find the chessboard corners
        ret, corners = cv2.findChessboardCorners(gray, (nx, ny), None)

        # If found, draw corners
        if ret == True:
            # Draw and display the corners
            objpoints.append(objp)
            imgpoints.append(corners)
            #cv2.drawChessboardCorners(img, (nx, ny), corners, ret)
            #plt.imshow(img)

    return objpoints, imgpoints

def load_imgs(imgs_paths):
    imgs =[]
    for path in imgs_paths:
        imgs.append(mpimg.imread(path))

    return np.array(imgs)
```

```python
# Obtain camera calibration using OpenCV Chessboard
nx = 9
ny = 6

cal_imgs_paths = glob.glob('./camera_cal/calibration*.jpg')

cal_imgs = load_imgs(cal_imgs_paths)

objpoints, imgpoints = get_obj_img_points (cal_imgs, nx, ny)

img_size = cv2.imread(cal_imgs_paths[0]).shape[0:2]
img_size = img_size [::-1]

ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, img_size,None,None)

with open('camera_cal.pickle', 'wb') as handle:
    pickle.dump([ret, mtx, dist, rvecs, tvecs], handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
# Remove distortion from images
def undistort(image, read=False):
    if read == True:
        #image = cv2.imread(image)
        image = mpimg.imread(image)
    with open('camera_cal.pickle', 'rb') as handle:
        ret, mtx, dist, rvecs, tvecs = pickle.load(handle)
    
    undist = cv2.undistort(image, mtx, dist, None, mtx)

    return undist
```

```python
def warp_image(img, display=False, read = True):
    if read:
        undist = undistort(img, read=True)
    else:
        undist = undistort(img, read=False) 

    image_width = undist.shape[1]
    image_height = undist.shape[0]
    
    # Source coordinates
    src = np.float32([
            [image_width * 0.4475, image_height * 0.65],
            [image_width * 0.5525, image_height * 0.65],
            [image_width * 0.2, image_height * 0.975],
            [image_width * 0.8, image_height * 0.975],])
    
    # Destination coordinates
    dst = np.float32([
            [image_width * 0.2, image_height * 0.025],
            [image_width * 0.8, image_height * 0.025],
            [image_width * 0.2, image_height * 0.975],
            [image_width * 0.8, image_height * 0.975],])
    
    # Get transformation and inverse transformation matrices
    M = cv2.getPerspectiveTransform(src, dst)
    M_inv = cv2.getPerspectiveTransform(dst, src)
    warped = cv2.warpPerspective(undist, M, img_size)
    
    if display:
        f, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 6))
        f.tight_layout()
        ax1.imshow(undist)
        ax1.set_title('Undistorted Image', fontsize=20)
        ax2.imshow(warped)
        ax2.set_title('Undistorted and Warped Image', fontsize=20)
        plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
    
    else:
        return warped, M, M_inv
    
```

```python
def mask_apply(img,low,high):
    # Takes in low and high values and returns mask
    mask = cv2.inRange(img, low, high)
    return mask

def gaussian_blur(img, kernel=5):
    # Function to smooth image
    blur = cv2.GaussianBlur(img,(kernel,kernel),0)
    return blur

def sobel_lims(img, orient=0, sobel_kernel=3, thresh=(0, 255)):
    # Calculate directional gradient
    # Apply threshold
    if orient==0:
        img_s = cv2.Sobel(img,cv2.CV_64F, 1, 0)
    else:
        img_s = cv2.Sobel(img,cv2.CV_64F, 0, 1)
    img_abs = np.absolute(img_s)
    img_sobel = np.uint8(255*img_abs/np.max(img_abs))
    
    binary_output = 0*img_sobel
    binary_output[(img_sobel >= thresh[0]) & (img_sobel <= thresh[1])] = 1
    return binary_output

def draw_lines_hough(img, lines, img_height, smoothing, color=[255, 0, 0], thickness=24):
    # reused this function from project #1
    # draws all the hough lines with a certain slope
    
    img_bonnet_offset = np.int(img_height-30)
    lines = np.array(lines)
    if np.count_nonzero(lines)>0:
        for line in lines:
            for x1,y1,x2,y2 in line:
                slope = (y2-y1)/(x2-x1)    #positive slope for the right side and negative for the left
                if slope <= -1.5 or slope >= 1.5: 
                    if (x1 >= (img.shape[0]//2) and x2 >= (img.shape[0]//2)) and (y2 <= (img_bonnet_offset) and y1 <= (img_bonnet_offset)):
                        cv2.line(img, (x1, y1), (x2, y2), [0,0,255], thickness)     #draw every line detected by hough algo

                    elif (x1 < (img.shape[0]//2) and x2 < (img.shape[0]//2)) and (y2 <= (img_bonnet_offset) and y1 <= (img_bonnet_offset)):    #similarly for the left
                        cv2.line(img, (x1, y1), (x2, y2), [0,0,255], thickness)     #draw every line detected by hough algo
        

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap, smoothing=True):    
    # Returns an image with hough lines drawn.
    # Reused from project #1

    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines_hough(line_img, lines, img.shape[0], smoothing)
    return line_img

```

```python
# Create binary thresholded images to isolate lane line pixels
def lane_detection(image, show=False, read=False):
    
    if read:
        undist = undistort(image, read=True)
    else:
        undist = undistort(image, read=False) 
    
    undist_blur = gaussian_blur(undist, kernel = 5)
    img_size = undist.shape

    #Convert img to HSV & HLS
    HSV_img = cv2.cvtColor(undist_blur, cv2.COLOR_RGB2HSV)
    HLS_img = cv2.cvtColor(undist_blur, cv2.COLOR_RGB2HLS)


    # change image perspective in the different colorspaces
    transf_img, M, M_inv = warp_image(undist_blur, display=False, read = False)
    transf_img_HSV, M, M_inv = warp_image(HSV_img, display=False, read = False)
    transf_img_HLS, M, M_inv = warp_image(HLS_img, display=False, read = False)

    # Extract colors from image
    img_white = mask_apply(transf_img_HLS, hls_white_low, hls_white_high)
    img_yellow = mask_apply(transf_img_HSV, hsv_yellow_low, hsv_yellow_high)

    # Combine white and yellow 
    white_yellow_mask = cv2.bitwise_or(img_white, img_yellow)
    
    # Now we will apply sobel operator on the L and S channels in a similar way as in the course test
    HLS_img = cv2.cvtColor(transf_img, cv2.COLOR_RGB2HLS)

    img_L_channel = HLS_img[:,:,1]
    img_S_channel = HLS_img[:,:,2]
    
    img_L_sobel_x = sobel_lims(img_L_channel, 0, 5, (50,220))
    img_L_sobel_y = sobel_lims(img_L_channel, 1, 5, (50,220))

    img_S_sobel_x = sobel_lims(img_S_channel, 0, 5, (50,255))
    img_S_sobel_y = sobel_lims(img_S_channel, 1, 5, (50,255))
    
    # Combine results again
    L_sobel = cv2.bitwise_or(img_L_sobel_x, img_L_sobel_y)
    S_sobel = cv2.bitwise_or(img_S_sobel_x, img_S_sobel_y)

    LS_sobel = cv2.bitwise_or(L_sobel, S_sobel)
    LS_sobel = gaussian_blur(LS_sobel,25)
    
    #Use Hough lines to "filter" the data (reused from project #1)
    rho = 3    #size of the grid to create the hough lines (polar coords)
    theta = np.pi/180    #size of the grid to create hough lines (polar coords)
    threshold = 55    #number of votes required
    min_line_len = 80    #min line length for it to be considered
    max_line_gap = 10    #max space in between parts of the line for it to be considered
    LS_sobel_pre_hough = LS_sobel
    LS_sobel = hough_lines(LS_sobel, rho, theta, threshold, min_line_len, max_line_gap, smoothing=True)    #img with the calculated hough lines
    LS_sobel = cv2.cvtColor(LS_sobel,cv2.COLOR_RGB2GRAY)
    
    white_yellow_mask_pre_hough = white_yellow_mask
    white_yellow_mask = hough_lines(white_yellow_mask, rho, theta, threshold, min_line_len, max_line_gap, smoothing=True)    #img with the calculated hough lines
    white_yellow_mask = cv2.cvtColor(white_yellow_mask,cv2.COLOR_RGB2GRAY)


    # Combine color masks and sobel operator into a binary img
    bin_img = np.zeros_like(LS_sobel)
    bin_img[(white_yellow_mask >= 0.5) | (LS_sobel >= 0.5)] = 1
    bin_img_color = cv2.cvtColor(255*bin_img, cv2.COLOR_GRAY2RGB)
    bin_img_color[:,:,0] = 0
    
    # create representation over the original image
    processed_img = cv2.addWeighted(transf_img, 1, bin_img_color, 0.25, 0)


    if show == True:
        # Plotting thresholded images
        f, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4, sharey='col', sharex='row', figsize=(20,6))
        f.tight_layout()
        
        ax1.set_title('Original Image', fontsize=12)
        ax1.imshow(undistort(image, read=True))
        
        ax2.set_title('Transformed Image', fontsize=12)
        ax2.imshow((transf_img).astype('uint8'))
        
        ax3.set_title('white and yellow masks', fontsize=12)
        ax3.imshow(white_yellow_mask_pre_hough, cmap='gray')
        
        ax4.set_title('white and yellow + hough', fontsize=12)
        ax4.imshow(white_yellow_mask, cmap='gray')
        
        ax6.set_title('white mask', fontsize=12)
        ax6.imshow(img_white, cmap='gray')

        ax5.set_title('yellow mask', fontsize=12)
        ax5.imshow(img_yellow, cmap='gray')
        
        #ax5.set_title('sobel mask', fontsize=12)
        #ax5.imshow(LS_sobel_pre_hough, cmap='gray')

        #ax6.set_title('Sobel mask + hough', fontsize=12)
        #ax6.imshow(LS_sobel, cmap='gray')
        
        ax7.set_title('Combined masks', fontsize=12)
        ax7.imshow(bin_img, cmap='gray')
        
        ax8.set_title('All together', fontsize=12)
        ax8.imshow(processed_img)
        
        plt.show()
        
    else: 
        return bin_img, img_white, img_yellow, white_yellow_mask, LS_sobel, undist, transf_img, M_inv
```

```python
def init_lines(img, steps, search_window, h_smoothing=7):
    x_points = []
    y_points = []
    masked_img = img[:, search_window[0]:search_window[1]]
    pixels_per_step = img.shape[0] // steps
    prev_highest_peak = 0
    line_mask = np.zeros_like(img)

    for i in range(steps):
        start = masked_img.shape[0] - (i * pixels_per_step)
        end = start - pixels_per_step
        hist = np.sum(masked_img[end:start, :], axis=0)
        hist_smooth = signal.medfilt(hist, h_smoothing)
        peaks = np.array(signal.find_peaks_cwt(hist_smooth, np.arange(1, 5)))

        highest_peak = highest_n_peaks(hist_smooth, peaks, n=1, threshold=5)
        if len(highest_peak) == 1:
            highest_peak = highest_peak[0]

            # outlyer removal
            if np.abs(highest_peak-prev_highest_peak)>=60 and i>0:
                highest_peak = prev_highest_peak
            else:
                prev_highest_peak = highest_peak
                    
            window_sz = 60
            line_mask[end:start,search_window[0]+highest_peak-window_sz:search_window[0]+highest_peak+window_sz] = 1.

            prev_highest_peak = highest_peak

    return line_mask

def highest_n_peaks(histogram, peaks, n=2, threshold=0):
    if len(peaks) == 0:
        return []

    peak_list = [(peak, histogram[peak]) for peak in peaks if histogram[peak] > threshold]
    peak_list = sorted(peak_list, key=lambda x: x[1], reverse=True)

    if len(peak_list) == 0:
        return []

    x, y = zip(*peak_list)
    x = list(x)

    if len(peak_list) < n:
        return x

    return x[:n]



def update_mask(img,poly_fit,window_sz):
    
    # This function returns masks for points used in computing polynomial fit. 
    mask_poly = np.zeros_like(img)
    img_size = np.shape(img)
    window_sz = np.int(window_sz)

    poly_pts = []
    pt_y_all = []
    original_window_sz = window_sz
    for i in range(8):
        window_sz = original_window_sz + i* original_window_sz//6
        img_y1 = np.int(img_size[0]-img_size[0]*i/8)
        img_y2 = np.int(img_size[0]-img_size[0]*(i+1)/8)

        pt_y = (img_y1+img_y2)/2
        pt_y_all.append(pt_y)
        poly_pt = np.int(np.round(poly_fit[0]*pt_y**2 + poly_fit[1]*pt_y + poly_fit[2]))
    
        poly_pts.append(poly_pt)
    
        mask_poly[img_y2:img_y1,poly_pt-window_sz:poly_pt+window_sz] = 1.     

    return mask_poly, np.array(poly_pts),np.array(pt_y_all)
    

def get_val(y,pol_a):
    # Returns value of a quadratic polynomial 
    return pol_a[0]*y**2+pol_a[1]*y+pol_a[2]

def lane_line_draw(img,pts,color):
    # This function draws lines connecting 10 points along the polynomial
    pts = np.int_(pts)
    for i in range(10):
        x1 = pts[0][i][0]
        y1 = pts[0][i][1]
        x2 = pts[0][i+1][0]
        y2 = pts[0][i+1][1]
        cv2.line(img, (x1, y1), (x2, y2),color,50)

def moving_average(a, n=3):
    # Moving average
    n = np.int(n)
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
        
def get_curvature(pol_a,y_pt):
    # Returns curvature of a quadratic
    A = pol_a[0]
    B = pol_a[1]
    R_curve = (1+(2*A*y_pt+B)**2)**1.5/2/A
    return R_curve

#def stack_arr(arr):
    # Stacks 1-channel array into 3-channel array to allow plotting
#    return np.stack((arr, arr,arr), axis=2)
```

```python
def process_video(image, ):
    
    global left_fit_prev   
    global right_fit_prev
    global color_left
    global color_right

    global right_line_mask
    global left_line_mask

    global mask_init
    
    global enable_diag

    # extract lane line info
    bin_img, img_white, img_yellow, white_yellow_mask, LS_sobel, undist, transf_img, M_inv = lane_detection(image, show=False, read=False)

    # if its the first operation do a blind search
    if mask_init == 0:
        bin_img = gaussian_blur(bin_img, 5)
        bin_img = gaussian_blur(bin_img,5)
        right_line_mask = init_lines(bin_img, 8, (bin_img.shape[1] // 2, bin_img.shape[1]), h_smoothing=3)
        left_line_mask = init_lines(bin_img, 8, (0, bin_img.shape[1] // 2), h_smoothing=3)

    # find pixels on the mask area
    right_pixels = cv2.bitwise_and(bin_img, right_line_mask)
    left_pixels = cv2.bitwise_and(bin_img, left_line_mask)
    
    right_points = np.argwhere(right_pixels>.5)
    left_points = np.argwhere(left_pixels>.5)
        
    # if there are not enough hits use previous data
    if len(right_points)<5:
        right_fit = right_fit_prev
        color_right = (255,0,0)
    else:
        all_x_R = right_points.T[0]
        all_y_R =right_points.T[1]
        right_fit = np.polyfit(all_x_R, all_y_R, 2) # fit polynomial to the points found
        if np.sum(cv2.bitwise_and(right_pixels, white_yellow_mask))>1000:
            color_right = (255,255,255)

    # similar for the left line
    if len(left_points)<5:
        left_fit = left_fit_prev
        color_left = (255,0,0)    # if lane "not found" paint it red
    else:
        all_x_L = left_points.T[0]
        all_y_L =left_points.T[1]
        left_fit = np.polyfit(all_x_L, all_y_L, 2)
        if np.sum(cv2.bitwise_and(left_pixels, white_yellow_mask))>1000:
            color_left = (255,255,0)    #if lane is found with enough confidence paint it yellow

    # store previous values
    if mask_init == 0:
        mask_init = 1
        right_fit_prev = right_fit
        left_fit_prev  = left_fit
       
    # Check the error between the current coefficient and the one from the previous frame
    err_p_R = np.sum((right_fit[0]-right_fit_prev[0])**2) 
    err_p_R = np.sqrt(err_p_R)
    if err_p_R>.001:
        right_fit = right_fit_prev
    else:
        right_fit = .075*right_fit+.925*right_fit_prev
        
    # Check the error between the current coefficient and the one from the previous frame
    err_p_L = np.sum((left_fit[0]-left_fit_prev[0])**2)
    err_p_L = np.sqrt(err_p_L)
    if err_p_L>.001:
        left_fit =  left_fit_prev
    else:
        left_fit =  .075* left_fit+.925* left_fit_prev
    

    # Calculate lane mask for the next frame 
    left_line_mask,left_pts,img_pts = update_mask(bin_img,left_fit,window_size)
    right_line_mask,right_pts,img_pts = update_mask(bin_img,right_fit,window_size)
     
        
    # Calculate the lanes
    right_y = np.arange(11)*img_size[0]/10
    right_fitx = right_fit[0]*right_y**2 + right_fit[1]*right_y + right_fit[2]

    left_y = np.arange(11)*img_size[0]/10
    left_fitx = left_fit[0]*left_y**2 + left_fit[1]*left_y + left_fit[2]
    
    warp_zero = np.zeros_like(bin_img).astype(np.uint8)
    color_warp = np.dstack((warp_zero, warp_zero, warp_zero))

    # Adapt x and y points into an easier format
    pts_left = np.array([np.transpose(np.vstack([left_fitx, left_y]))])
    pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx, right_y])))])
    pts = np.hstack((pts_left, pts_right))
    

    # Calculate intersects
    left_bot = get_val(img_size[0],left_fit)
    right_bot = get_val(img_size[0],right_fit)
    
    # Calculate center of the lane (in the bottom)
    val_center = (left_bot+right_bot)/2.0
    
    # Calculate lane offset
    dist_offset = val_center - undist.shape[1]/2
    dist_offset = np.round(dist_offset/1.97,2)
    str_offset = 'Lane center offset: ' + str(dist_offset) + ' cm.'
    
    if dist_offset>50:
        cv2.fillPoly(color_warp, np.int_([pts]), (220,0, 100))    # if lateral offset is more than 50cm paint the lane red
    else:
        cv2.fillPoly(color_warp, np.int_([pts]), (5, 240, 140))   # else paint it green
        

    # Draw the lane lines   
    lane_line_draw(color_warp,np.int_(pts_left),color_left)
    lane_line_draw(color_warp,np.int_(pts_right),color_right)
    # warp back the image to its original form using the inverse transformation matrix
    img_inv_transformed = cv2.warpPerspective(color_warp, M_inv, (undist.shape[1], undist.shape[0])) 

    # create representation of the result warped
    transformed_processed = cv2.addWeighted(transf_img, 1, color_warp, 0.15, 0)

    # Combine the result with the original image
    processed_img = cv2.addWeighted(undist, 1, img_inv_transformed, 0.5, 0)
    
    # calculate line curvatures
    Left_curve = get_curvature(left_fit,img_size[0]/2)
    Right_curve = get_curvature(right_fit,img_size[0]/2)
    
    if Left_curve > 7500 or Right_curve > 7500 or Left_curve < -7500 or Right_curve < -7500:
        str_curv = 'Curvature: almost straight' 

    else:
        str_curv = 'Curvature: Left = ' + str(np.round(Left_curve,2)) + ', Right = ' + str(np.round(Right_curve,2)) 

    font = cv2.FONT_HERSHEY_SIMPLEX    
    cv2.putText(processed_img, str_curv, (30, 60), font, 1, (220, 175, 40), 2)
    cv2.putText(processed_img, str_offset, (30, 90), font, 1, (220, 175, 40), 2)

    
    right_fit_prev = right_fit
    left_fit_prev  = left_fit

    
    
    if enable_diag:
        diagFrame = np.zeros((1080, 1920, 3), dtype=np.uint8)
        diagFrame[0:720, 0:1280] = processed_img
        diagFrame[0:240, 1280:1600] = cv2.resize(transf_img, (320,240), interpolation=cv2.INTER_AREA) 
        #diagFrame[0:240, 1280:1600] = cv2.resize(cv2.cvtColor(255*LS_sobel_pre_hough, cv2.COLOR_GRAY2RGB), (320,240), interpolation=cv2.INTER_AREA) 
        diagFrame[0:240, 1600:1920] = cv2.resize(cv2.cvtColor(255*LS_sobel, cv2.COLOR_GRAY2RGB), (320,240), interpolation=cv2.INTER_AREA)
        diagFrame[240:480, 1280:1600] = cv2.resize(cv2.cvtColor(img_yellow, cv2.COLOR_GRAY2RGB), (320,240), interpolation=cv2.INTER_AREA)
        diagFrame[240:480, 1600:1920] = cv2.resize(cv2.cvtColor(img_white, cv2.COLOR_GRAY2RGB), (320,240), interpolation=cv2.INTER_AREA)
        diagFrame[600:1080, 1280:1920] = cv2.resize(transformed_processed, (640,480), interpolation=cv2.INTER_AREA)
        #diagFrame[600:1080, 1280:1920] = cv2.resize(color_warp, (640,480), interpolation=cv2.INTER_AREA)*4
        diagFrame[720:960, 0:320] = cv2.resize(cv2.cvtColor(255*white_yellow_mask, cv2.COLOR_GRAY2RGB), (320,240), interpolation=cv2.INTER_AREA)

        diagFrame[720:960, 320:640] = cv2.resize(cv2.cvtColor(255*bin_img, cv2.COLOR_GRAY2RGB), (320,240), interpolation=cv2.INTER_AREA)
        diagFrame[720:960, 640:960] = cv2.resize(cv2.cvtColor(255*left_line_mask+255*right_line_mask, cv2.COLOR_GRAY2RGB), (320,240), interpolation=cv2.INTER_AREA)
        diagFrame[720:960, 960:1280] = cv2.resize(cv2.cvtColor(255*cv2.bitwise_and(bin_img,bin_img, mask=left_line_mask+right_line_mask), cv2.COLOR_GRAY2RGB), (320,240), interpolation=cv2.INTER_AREA)
        
        return diagFrame
    
    else:
        return processed_img
```
